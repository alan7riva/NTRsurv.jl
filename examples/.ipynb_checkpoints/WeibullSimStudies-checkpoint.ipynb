## NTR distributions for Kaplan-Meier alternative

We say that a random probability distribution supported on $\mathbb{R}^+$, with cumulative distribution function $F$, is neutral to the right if for positive real numbers $0\leq t_1<\ldots <t_n$ the random vector

$$
(V_1,\ldots,V_n)=\left( F(t_1),\frac{F(t_2)-F(t_1)}{1-F(t_1)}, \ldots , \frac{F(t_n)-F(t_{n-1})}{1-F(t_{n-1})}\right)
$$

has mutually independent entries.

Such distributions can be characterized by having survival function 

$$
S(t)=1-F(t)=\exp\left(-Y(t) \right)
$$

where $Y(t)$ is an almost surely non-decreasing function with independent increments, known in the stochastic processes literature as a subordinator. The so called subordinator $Y$ is in turn characterized by a Poisson process $N$ with intensity $\nu(\mathrm{d}s,\mathrm{d}x)$ in $\mathbb{R}^+\times\mathbb{R}^âº$, the "s" axis corresponding to weights in the jumps of $Y$ while the "x" axis corresponds to locations of those jumps, such that we can express

$$
Y(t)=\int_0^\infty \int_0^t N(s,x)\mathrm{d}x\mathrm{d}s.
$$

We say that a random variable with survival function $S$ as above has a $\text{NTR}(Y)$ distribution.

**NTR with Gamma processes workflow for survival analysis**

If the intensity $\nu$ has the form

$$
\nu(\mathrm{d}s,\mathrm{d}x)  = \beta \frac{e^{-\alpha s}}{s} \mathrm{d}s \mathrm{d} \kappa(x)
$$

with $\alpha,\beta>0$ we say that the associated subordinator is a Gamma process as it can be checked that

$$
Y(t)\sim \text{Gamma}(\text{shape}=\beta\kappa(t),\text{ rate}=\alpha)
$$

The Laplace exponent of a Gamma process is given by
\begin{align*}
\psi_t(\lambda) =  \beta \log\left( 1 +\frac{\lambda}{\alpha} \right) \kappa(t).
\end{align*}

We choose $$\beta = \frac{1}{\log\left( 1+\frac{1}{\alpha} \right)}$$ with which

$$
\mathbb{E}\left[ S(t) \right]=e^{-\psi_t(1)} = e^{ -\kappa(t) },
$$

so $\kappa$ serves as a cuulative hazard baseline for centering the model a priori.

On the other hand
$$
\mathbb{E}\left[ (S(t))^2 \right]=e^{-\psi_t(2)} = exp\left( -\frac{\log\left(1+\frac{2}{\alpha} \right)}{\log\left(1+\frac{1}{\alpha} \right)} \kappa(t)   \right)
$$
and

$$
\text{Var}\left( S(t) \right) = exp\left( -\frac{\log\left(1+\frac{2}{\alpha} \right)}{\log\left(1+\frac{1}{\alpha} \right)} \kappa(t)   \right) - e^{-2\kappa(t)}.
$$

We observe that

$$
\frac{\log\left(1+\frac{\lambda}{\alpha} \right)}{\log\left(1+\frac{1}{\alpha} \right)}
=\frac{\log\left(\alpha + \lambda  \right) - \log(\alpha)}{ \log\left(\alpha+ 1 \right) - \log(\alpha)}
= \frac{ \frac{ \log\left(\alpha + \lambda  \right) }{\log(\alpha)} - 1}{ \frac{ \log\left(\alpha+ 1 \right) }{\log(\alpha)} - 1 }
$$

so

$$
\lim\limits_{\alpha \to 0} \frac{\log\left(1+\frac{\lambda}{\alpha} \right)}{\log\left(1+\frac{1}{\alpha} \right)}=1
$$

and by de L'Hopital's rule

\begin{align*}
\lim\limits_{\alpha \to \infty} \frac{\log\left(1+\frac{\lambda}{\alpha} \right)}{\log\left(1+\frac{1}{\alpha} \right)} &= \lim\limits_{\alpha \to \infty} \frac{ \log(\alpha+\lambda)-\log(\alpha) }{ \log(\alpha+1)-\log(\alpha) }=
\lim\limits_{\alpha \to \infty } \frac{ (\alpha+\lambda)^{-1}  -  \alpha^{-1}  }{ (\alpha+1)^{-1} - \alpha^{-1} }
=  \lambda \lim\limits_{\alpha \to \infty} \frac{ (\alpha + 1) }{ ( \alpha+\lambda ) } = \lambda.
\end{align*}

```julia
# Check limit behavior
Î» = 2.0
h(s) = log(1+Î»/s)/log(1+1/s)
t=collect(0.0:1.5:1000.0)
plot(t,f.(t),label=L"h(x)=\beta \frac{ \log\left(1 + \frac{\lambda}{x} \right) }{ \log\left(1 + \frac{1}{x} \right) }",legend=:right)
hline!([Î»],linestyle=:dash,label=latexstring("$(Int(Î»))"))
```

It also follows that

$$
\lim\limits_{\alpha \to 0}\text{Var}\left( S(t) \right)= e^{-\kappa(t)}- e^{-2\kappa(t)}
$$

and

$$
\lim\limits_{\alpha \to \infty} \text{Var}\left( S(t) \right) = e^{-2\kappa(t)}- e^{-2\kappa(t)}=0.
$$

For a Bayesian analysis we can work with the reparaetrization $\tilde{\alpha}=\frac{1}{\alpha}$ as it is simpler to set a prior on $\alpha$ which concentrates around $0$. Alternatively, in the proposed framework $\alpha$ is fixed and should reflect the a priori believes on the variance of the centering baseline survival.

We choose a priori an exponential baseline and variance modulating hyper-parameter for setting a NTR model.

```julia
# Create NTRsurv exponential baseline
baseline = ExponentialBaseline(1.0)
# Set variance modulation hyper-parameter
Î±â‚€ = 5.0
# Auxilliary quantities for plotting
Î²â‚€ = 1.0/log(1.0+1.0/Î±â‚€)
câ‚€ = Î²â‚€/Î±â‚€;
```

### Posterior characterization (conjugacy structure) for NTR priors

For possibly censored to the right survival data of size $n>1$

$$
\pmb{\mathcal{D}} = \left\{ T_i, \delta_i \right\}_{i=1}^n
$$

with 

$$
T_i = \min(X_i,C_i), \qquad 
\delta_i = \mathbf{1}\{X_i \le C_i\}, \qquad i=1,\dots,n,
\}
$$

where $X_1,\dots,X_n$ are i.i.d. nonnegative random variables representing true event times of interest and $C_1,\dots,C_n$ are i.i.d. nonnegative censoring times, independent of the $X_i$'s. We denote the distinct order statistics for observed survival data $\left\{ T_i=t_i\right\}_{i=1}^n$ as
$0 < t_{(1)} < t_{(2)} < \cdots < t_{(k_n)}$, with $k_n $ the observed number of distinct values. At each event time $t_{(k)}$ we define the number of exact observations at time $t_{(k)}$
$$
d_k = \sum_{i=1}^n \mathbf{1}\left[ \delta_i=1,\;T_i =  t_{(k)}\right\},
$$
the number of individuals at risk just prior to 
$t_{(k)}$
$$
r_k = \sum_{i=1}^n \mathbf{1}\{ T_i \ge t_{(k)}\}.
$$
and the number of individuals at risk just prior to 
$t_{(k)}$ excluding exact observations at $t_{(k)}$
$$
r^\star_k = r_k - d_k 
$$
We recall that the Kaplan--Meier estimator of the survival function $S(t)$ is defined as
$$
\widehat{S}_{KM}(t)
= \prod_{k \, : \, t_{(k)} \le t}
\left(1 - \frac{d_k}{r_k}\right),
\qquad t \ge 0,
$$
with the convention that an empty product equals 1.

Modeling with a $\text{NTR}(Y)$ prior where the subordinator $Y$ has an intensity of the form $\nu(\mathrm{d}s,\mathrm{d}x)=\nu(s,\mathrm{d}x)\mathrm{d}s$. We denote $\eta_t(s)=\nu(s,(0,t])$  If 
$$
\eta_{t_0}'(s)=\frac{\partial \eta_t(s)}{\partial t}\Bigg|_{t=t_0}$$ exists for all $t_0\in(0,t_f]$, $t_{(k_n)}<t_f$, then, restricted to $[0,t_f]$, the posterior distribution given survival data $\pmb{\mathcal{D}}$ is $\text{NTR}(\tilde{Y}_{\pmb{\mathcal{D}}})$ with

$$
\tilde{Y}_{\pmb{\mathcal{D}}}(t) = Y^\circ(t) + \sum_{ \left\{ k \, :\, d_k \geq 1 \right\} } W_k \delta_{T_{(k)}}
$$

where, setting $t_{(k_n+1)}=t_f$, $Y^\circ$ is a subordinator with intensity

$$
\nu^\circ (\mathrm{d}s,\mathrm{d}x) = \sum_{k=1}^{k_n+1} e^{- r_k s }\pmb{1}_{\left\{ x\in (t_{(k-1)},t_{(k)}] \right\}}
\nu(s,\mathrm{d}x)\mathrm{d}s,
$$

the random jump weights $W_k$, corresponding to a jump at fixed location $T_{(k)}$, have probability density function

\begin{align*}
f_k(s) = \frac{ e^{- \tilde{r}_k s}\left( 1-e^{-s} \right)^{d_k} \eta_{t_{(k)}}'(s)   }{ \int_0^\infty e^{- \tilde{r}_k u}\left( 1-e^{-u} \right)^{d_k} \eta_{t_{(k)}}'(u) \mathrm{d}u }
\end{align*}

and $\left\{ W_k \right\}_{k=1}^{k_n}\perp Y^\circ$.

Under the simplifyng assumption of $\nu$ being homogeneous and such that

$$
\nu(s,(0,t]) = \rho(s)\kappa(t)
$$

we have that

\begin{align*}
f_k(s) = \frac{ e^{- \tilde{r}_k s}\left( 1-e^{-s} \right)^{d_k} \rho(s)  }{ \int_0^\infty e^{- \tilde{r}_k u}\left( 1-e^{-u} \right)^{d_k} \rho ( u ) \mathrm{d}u }.
\end{align*}

**Posterior characterization with Gamma processes workflow for survival analysis**

Furthermore choosing $Y$ to be an homogeneous Gamma process we obtain further simplifications in that

$$
\nu^\circ (\mathrm{d}s,\mathrm{d}x) \Big|_{x \in (t_{(k-1)},t_{(k)}]}  =  \beta \frac{e^{-(\alpha + r_k) s}}{s} \mathrm{d}s \mathrm{d} \kappa(x),
$$

is a Gamma process intensity, with updated parameter $\alpha$ to $\alpha+r_k$ and the jump weight densities are given by

\begin{align*}
f_k(s) = \frac{ e^{-\left( \alpha + r^\star_k \right)s}\left( 1-e^{-s} \right)^{d_k}s^{-1} }{ \int_0^\infty e^{-\left( \alpha + r^\star_k \right)u}\left( 1-e^{-u} \right)^{d_k}u^{-1}\mathrm{d}u }.
\end{align*}

### Likelihood computation for NTR models

Computation of the likelihood for NTR models with survival data is is provided by

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \psi_{t_{(k)}}\left( r_k  \right) - \psi_{t_{(k-1)}}\left(r_k \right)   \right] \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \int_0^\infty e^{-r^\star_k s}\left(1-e^{-s}\right)^{d_k}
\eta'_{t_{(k)}}(s)\mathrm{d}s
\end{align*}

Under the simplifyng assumption of $\nu$ being homogeneous and such that $\nu(s,(0,t]) = \rho(s)\kappa(t)$ we have that

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right]\int_0^\infty (1-e^{-r_k s})\rho(s)\mathrm{d}s   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \kappa'\left( t_{(k)} \right) \int_0^\infty e^{-r^\star_k s}\left(1-e^{-s}\right)^{d_k}
 \rho(s)\mathrm{d}s
\end{align*}

**Likelihood computation with Gamma processes workflow for survival analysis**

Using the Laplace exponent formula for Gamma processes and doing a binomial expansion

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{r_k}{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \beta \kappa'\left( t_{(k)} \right) \sum_{j=0}^{d_k-1}\binom{d_k-1}{j}(-1)^{j} 
\int_0^\infty \left(1-e^{-s}\right) \frac{e^{-(\alpha +r^\star_k + j ) s}}{s}\mathrm{d}s 
\\ &= \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{r_k}{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \beta \kappa'\left( t_{(k)} \right) \sum_{j=0}^{d_k-1}\binom{d_k-1}{j}(-1)^{j}
\log\left(1 + \frac{ 1 }{\alpha +r^\star_k + j } \right)
\end{align*}

In the particular case that there are no repetitions in the exact observations, i.e. $d_k=1$ for $1\leq k \leq k_n$, then no use of a binomial expansion is necessary and we obtain the simplification

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{r_k}{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} }\beta \kappa'\left( t_{(k)} \right) \log\left(1 + \frac{1}{\alpha+ r^\star_k} \right)
\end{align*}

### Posterior mean survival for NTR models

From the posterior characterization we can obtain the posteror mean survival function which can be used as an estimator

\begin{align*}
\hat{S}(t) &= \mathbb{E}\left[  \mathbb{P}\left[ X >t \,|\, Y,\, \pmb{\mathcal{D}}\right] \,|\, \pmb{\mathcal{D}}\ \right]
= \mathbb{E}\left[e^{-Y^\circ(t) }  \,|\, \pmb{\mathcal{D}}\  \right] \prod_{ \{ k \,:\, d_k \geq 1,\, t_{(k)} \leq t \} } \mathbb{E}\left[e^{-W_k} \,|\, \pmb{\mathcal{D}}\  \right]
\\ &= e^{-\sum_{k=1}^{k_n+1}\left( \psi^\circ_{t\wedge t_{(k)}}(1) - \psi^\circ_{t\wedge t_{(k-1)}}(1)  \right) }
\prod_{ \{ k \,:\, d_k \geq 1,\, t_{(k)} \leq t \} }\frac{ \int_0^\infty e^{-( r^\star_k + 1)s}\left(1-e^{-s} \right)^{d_k}\eta'_{t_{(k)}}(s)\mathrm{d}s }{ \int_0^\infty e^{- r^\star_ks}\left(1-e^{-s} \right)^{d_k}\eta'_{t_{(k)}}(s)\mathrm{d}s }
\\ &= e^{-\sum_{k=1}^{k_n+1}\left( \psi^\circ_{t\wedge t_{(k)}}(1) - \psi^\circ_{t\wedge t_{(k-1)}}(1)  \right) }
\prod_{ \{ k \,:\, d_k \geq 1,\, t_{(k)} \leq t \} }\frac{ \int_0^\infty e^{-( r^\star_k + 1)s}\left(1-e^{-s} \right)^{d_k}\eta'_{t_{(k)}}(s)\mathrm{d}s }{ \int_0^\infty e^{- r^\star_ks}\left(1-e^{-s} \right)^{d_k}\eta'_{t_{(k)}}(s)\mathrm{d}s }
\end{align*}

where $t_{(k_n+1)}=\infty$ and $\psi^\circ_t(\lambda)= -\log\left( \mathbb{E}\left[ e^{-\lambda Y^\circ(t)} \right] \right)$ is the Laplace exponent of $Y^\circ(t)$.

Again, under the case of $\nu$ being homogeneous and such that $\nu(s,(0,t]) = \rho(s)\kappa(t)$ we have the simplification

\begin{align*}
\hat{S}(t) &=  e^{-\sum_{k=1}^{k_n+1}\left( \kappa( t\wedge t_{(k)} ) - \kappa( t\wedge t_{(k-1)} )  \right) \int_0^\infty e^{-r_k s} (1-e^{- s})\rho(s)\mathrm{d}s }
\prod_{ \{ k \,:\, d_k \geq 1,\, t_{(k)} \leq t \} }\frac{ \int_0^\infty e^{-( r^\star_k + 1)s}\left(1-e^{-s} \right)^{d_k} \rho(s)\mathrm{d}s }{ \int_0^\infty e^{- r^\star_ks}\left(1-e^{-s} \right)^{d_k}\rho(s)\mathrm{d}s };
\end{align*}

and for the Gamma process case in particular we get

\begin{align*}
\hat{S}(t) &=  e^{-\sum_{k=1}^{k_n+1}\left( \kappa( t\wedge t_{(k)} ) - \kappa( t\wedge t_{(k-1)} )  \right) \log\left(1 +\frac{1}{\alpha+r_k} \right)}
\prod_{ \{ k \,:\, d_k \geq 1,\, t_{(k)} \leq t \} }\frac{ \sum_{j=0}^{d_k-1}\binom{d_k-1}{j}(-1)^{j-1}
\log\left( \frac{\alpha +r^\star_k + j +1 }{\alpha +r^\star_k + j + 2 } \right)
 }{ \sum_{j=0}^{d_k-1}\binom{d_k-1}{j}(-1)^{j-1}
\log\left( \frac{\alpha +r^\star_k + j }{\alpha +r^\star_k + j + 1 } \right) };
\end{align*}

We illustrate the above estimator for synthetic Weibull data.

### Posterior simulation of NTR survival functions

From the posterior characterization of the NTR model, we have that simulation of posterior survival draws is feasible by drawing

$$
\tilde{S}_{\pmb{\mathcal{D}}}(t) = \exp\left( -\tilde{Y}_{\pmb{\mathcal{D}}}(t) \right) 
$$

It is seen from the posterior characterization that in order to draw simulations of $\tilde{Y}_{\pmb{\mathcal{D}}}$ in $[0,t_f]$ we only need to:

1) Draw Gamma processes in $(t_{(k-1)},t_{(k)}]$ with parameters $\kappa$, $\beta_\alpha$ and $\alpha^\circ=\alpha+r_k$.

2) Draw random weights, corresponding to fixed jump location $t_{(k)}$,  $W_k\sim\mathcal{L}(f_k)$.

For the random weights in our Gamma process workflow we use an efficient acceptance-rejection sampler in the sense that the probability of acceptance goes to one as $r_k^\star$ goes to infinity, provided that $\frac{1}{n}\sum_{i=1}^n \delta_i$ converges to a positive limit. 

##### Case with no repetitions
Let the kernel of $f_k$ be denoted by
$$
f_k^{\text{ker}}= e^{-\left( \alpha + r^\star_k \right)s}\left( 1-e^{-s} \right)^{d_k}s^{-1}  
$$
Use proposal distribution $\text{Exp}(\alpha+r_k^\star)$ with corresponding density denoted $g$. Accept a draw $V$ from $g$ if for an independent $U\sim\text{Uniform}$
$$
\frac{1-e^{-V}}{V}\leq U.
$$ 

##### Case with repetitions

Use proposal distribution $\text{Gamma}(d_k,\alpha + r_k^\star)$ with corresponding density denoted $g$. Accept a draw $V$ from $g$ if for an independent $U\sim\text{Uniform}$
$$
\left( \frac{1-e^{-V}}{V} \right)^{d_k}\leq U.
$$ 

Such posterior simulation is convenient for Monte-Carlo computation of credible bands, as provided with the function `cred_band`.

```julia
NTR_band_d, NTR_band_m, NTR_band_u = cred_band(0.05,3000,t,model_NTR)
```

```julia
# Plot for posterior credible band 
post_draws_plot = plot(t, NTR_band_m,  ribbon = (NTR_band_m .- NTR_band_d, NTR_band_u .- NTR_band_m), xlabel=L"t", ylabel=L"S(t)", fillalpha=0.3, label="NTR posterior mean survival \n with 95% credible band",size=(600,400))
plot!( t, ccdf.( Weibull(k,Î»), t), color=:black,linestyle=:dash, label="True survival")
plot!(km.events.time,km.survival,seriestype = :steppost,label="Kaplan-Meier estimator")
```

## Cox NTR distributions

We can consider the inclussion of covariates in NTR based models. Denote such covariates as $\pmb{X}$ for $T$ a possibly censored to the right random variable. The Cox regression model for NTR Cox-baseline is determined by covariate conditional survival function
$$
S(t;\pmb{X})=\exp\left(-g(\pmb{b},\pmb{X})Y(t) \right)
$$

### Posterior characterization (conjugacy structure) for Cox NTR priors

For the Cox regression setting we define

\begin{align*}
h^{(e)}_k(\pmb{b}) &= \sum_{\{i\,:\, t_i = t_{(k)},\, \delta_i=1 \}} g(\pmb{b},\pmb{X}_i), \quad \bar{h}^{(c)}_k(\pmb{b}) = \sum_{ \{i\,:\, t_i = t_{(k)},\, \delta_i=0 \} }g(\pmb{b},\pmb{X}_i)
\\ 
\bar{h}^{(e)}_k(\pmb{b}) &= \sum_{i=k}^{k_n} h^{(e)}_i(\pmb{b},\hat{\pmb{X}}), \quad \hphantom{} \bar{h}^{(c)}_k(\pmb{b}) = \sum_{i=k}^{k_n} h^{(c)}_i(\pmb{b},\hat{\pmb{X}}).
\end{align*}
So the quantities related to the numbers of at risk patients become
\begin{align*}
h_k(\pmb{b})=\bar{h}^{(c)}_k(\pmb{b})+\bar{h}^{(e)}_k(\pmb{b}), &\quad h^\star_k(\pmb{b})=\bar{h}^{(c)}_k(\pmb{b})+\bar{h}^{(e)}_{k+1}(\pmb{b})
\end{align*}
where $\bar{h}^{(e)}_{k_n+1}(\pmb{b})=0$.

Consider a posteriori the survival time $T^{(\text{new})}$ of a new patient with corresponding covariate vector $\pmb{X}^{(\text{new})}$. With such notation, the posterior distribution of $f(\pmb{\beta},\pmb{X}^{(\text{new})})Y$ in $(0,t_f]$ is given by

$$
Y^\circ(t) + \sum_{ \left\{ k \, :\, d_k \geq 1 \right\} } W_k \delta_{T_{(k)}}
$$

where $Y^\circ$ is a subordinator with intensity

$$
\nu^\circ (\mathrm{d}s,\mathrm{d}x) = \sum_{k=1}^{k_n+1} e^{-\left( \frac{  h_k(\pmb{b}) }{ g(\pmb{b},\pmb{X}^{(\text{new})})}\right)s }\pmb{1}_{\left\{ x\in (t_{(k-1)},t_{(k)}] \right\}}
\nu(s,\mathrm{d}x)\mathrm{d}s, 
$$

the random jump weights $W_k$, corresponding to a jump at fixed location $T_{(k)}$, have probability density function
$W_j\sim\mathcal{L}(f_j)$ with

\begin{align*}
f_k(s) = \frac{ e^{-\left( \frac{ h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)s}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \} }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right) s} \right) \eta_{t_{(k)}}'(s)  }{ \int_0^\infty e^{-\left( \frac{ h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \} }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u} \right) \eta_{t_{(k)}}'(u) \mathrm{d}u }.
\end{align*}

and $\left\{ W_k \right\}_{k=1}^{k_n}\perp Y^\circ$.

Under the simplifyng assumption of $\nu$ being homogeneous and such that

$$
\nu(s,(0,t]) = \rho(s)\kappa(t)
$$

we have that

\begin{align*}
f_k(s) = \frac{ e^{-\left( \frac{ h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)s}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \} }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right) s} \right) \rho(s)  }{ \int_0^\infty e^{-\left( \frac{ h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \} }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u} \right) \rho(u) \mathrm{d}u }.
\end{align*}

**Posterior characterization with Gamma processes workflow for survival analysis**

Furthermore choosing the Cox-baseline $Y$ to be an homogeneous Gamma process we obtain further simplifications in that

$$
\nu^\circ (\mathrm{d}s,\mathrm{d}x) \Big|_{x \in (t_{(k-1)},t_{(k)}]}  =  \beta \frac{e^{-\left(\frac{\alpha + h_k(\pmb{b})}{ g\left(\pmb{b},\pmb{X}^{(\text{new})} \right) } \right) s}}{s} \mathrm{d}s \mathrm{d} \kappa(x),
$$

is a Gamma process intensity, with updated parameter $\alpha$ to $(\alpha+h_k)/g\left(\pmb{b},\pmb{X}^{(\text{new})} \right)$ and the jump weight densities are given by

\begin{align*}
f_k(s) = \frac{ e^{-\left( \frac{ \alpha + h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)s}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \}  }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right) s} \right)s^{-1} }{ \int_0^\infty e^{-\left( \frac{ \alpha + h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u}\prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \} }\left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u} \right)u^{-1}\mathrm{d}u }.
\end{align*}

In case that there are no reptitions of exact observations we obtain

\begin{align*}
f_k(s) = \frac{ e^{-\left( \frac{ \alpha + h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)s} \left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right) s} \right)s^{-1} }{ \int_0^\infty e^{-\left( \frac{ \alpha + h_k(\pmb{b}) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u}  \left( 1-e^{-\left(\frac{ g(\pmb{b},\pmb{X}_l) }{g(\pmb{b},\pmb{X}^{(\text{new})})} \right)u} \right)u^{-1}\mathrm{d}u }.
\end{align*}

### Likelihood computation for Cox NTR models

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \psi_{t_{(k)}}\left( h_k(\pmb{b})  \right) - \psi_{t_{(k-1)}}\left( h_k(\pmb{b})  \right)   \right] \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \int_0^\infty e^{-h^\star_k(\pmb{b})  s} \prod_{ \{ l \,:\, t_l= t_{(k)}\, , \delta_l = 1 \}  }\left(1-e^{-g(\pmb{b},\pmb{X}_l)s}\right)
\eta'_{t_{(k)}}(s)\mathrm{d}s
\end{align*}

In case that there are no reptitions of exact observations we obtain the simplified formula

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \psi_{t_{(k)}}\left( h_k(\pmb{b})   \right) - \psi_{t_{(k-1)}}\left( h_k(\pmb{b})  \right)   \right] \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \int_0^\infty e^{-h^\star_k(\pmb{b})  s} \left(1-e^{-g(\pmb{b},\pmb{X}_k)s}\right)
\eta'_{t_{(k)}}(s)\mathrm{d}s
\end{align*}

**Likelihood computation with Gamma processes workflow for survival analysis**

Using the Laplace exponent formula for Gamma processes and denoting $
\underline{g}_k = \min_\left\{ l \,:\, t_l = t_{(k)},\, \delta_l = 1 \right\}g(\mathbf{b}, \mathbf{X}_l) 
$ with corresponding $\underline{l}_k = \arg \min_\left\{ l \,:\, t_l = t_{(k)},\, \delta_l = 1 \right\}g(\mathbf{b}, \mathbf{X}_l) 
$ and $S_k = \left\{ l \,:\, t_l = t_{(k)},\, \delta_l = 1 \right\} \setminus \underline{l}_k$

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{h_k(\pmb{b})  }{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \beta \kappa'\left( t_{(k)} \right) \sum_{C\subset S_k }(-1)^{\# C} \int_0^\infty e^{-\left( h^\star_k(\pmb{b})  +  \sum_{l\in C}g(\pmb{b},\pmb{X}_l)\right)s}(1-e^{- g(\pmb{b},\pmb{X}_{\underline{l}_k} )s }) e^{-\alpha s}{s}\mathrm{d}s
\\ &= \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{h_k(\pmb{b})}{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} } \beta \kappa'\left( t_{(k)} \right) \sum_{C\subset S_k }(-1)^{\# C} \log\left(1 +\frac{ g(\pmb{b},\pmb{X}_{\underline{l}_k})}{\alpha + h^\star_k(\pmb{b})  + \sum_{l\in C}g(\pmb{b},\pmb{X}_l) } \right)
\end{align*}

In the particular case that there are no repetitions in the exact observations, i.e. $d_k=1$ for $1\leq k \leq k_n$, then no use of a binomial expansion is necessary and we obtain the simplification

\begin{align*}
l(\pmb{c},\pmb{D}) & = \exp\left( -\sum_{k=1}^{k_n} \left[ \kappa\left( t_{(k)} \right) - \kappa\left( t_{(k-1)}  \right)   \right] \beta \log\left( 1 + \frac{h_k(\pmb{b})}{\alpha} \right)   \right)
\prod_{ \{ k \,:\, d_k \geq 1 \} }\beta \kappa'\left( t_{(k)} \right) \log\left(1 + \frac{g(\pmb{b},\pmb{X}_k)}{\alpha+ h^\star_k(\pmb{b})} \right)
\end{align*}

```julia
function SuffStatsRegreNTR(c::Vector{Float64},data::DataRegreNTRrep,f::Function)
    m = data.m
    Záµ‰ = copy(data.Záµ‰)
    Zá¶œ = copy(data.Zá¶œ)
    l = [  isempty(Záµ‰[i]) ? nothing : findmin([ f(c,v) for v in Záµ‰[i] ])[2] for i in 1:m ]
    Î½ = [ isnothing(i) ? 0.0 : f(c,Záµ‰[i]) for i in l ]
    for i in 1:m
        deleteat!( Záµ‰[i], l[i] )
    end
    háµ‰ = [ sum( [ f(c,v) for v in Záµ‰[i] ], init=0.0) for i in 1:m ] # frequencies of exact bservations
    há¶œ = [ sum( [ f(c,v) for v in Zá¶œ[i] ], init=0.0) for i in 1:m ] # frequencies of censored observations
    Háµ‰ = [ cumsum( háµ‰[end:-1:1] )[end:-1:1]; 0]
    Há¶œ = [ cumsum( há¶œ[end:-1:1] )[end:-1:1]; 0]
    Râ‚ = Háµ‰ .+ Há¶œ 
    Râ‚‚ = Há¶œ .+ [ Háµ‰[2:end]; 0]
    F = [ [ [ length(v), sum( [ f(c,z) for z in Záµ‰[k][v]], init=0.0)] for v in collect(subsets(1:length(Záµ‰[k]))) ] for k in 1:m ]
    return Râ‚, Râ‚‚, Î½, F
end
```

```julia
function loglikRegreNTR(c::Vector{Float64},Î±::Real,baseline::BaselineNTR,data::DataRegreNTRrep)
    l = 0.0
    Îº = baseline.Îº
    dÎº = baseline.dÎº
    Î² = 1.0/log(1.0+1.0/Î±)
    m = data.m
    X =  [0.0;data.T]
    Râ‚, Râ‚‚, Î½, F = SuffStatsRegreNTR(c,data,f)
    náµ‰ = data.náµ‰
    cont_incr(k::Int64) = Î²*( Îº(X[k+1])-Îº(X[k]) )*log( Î±/(Î± + Râ‚[k]) )    
    disc_incr(k::Int64) = log( dÎº(X[k+1]) ) + log(Î²) + log( sum( [ (-1.0)^v[1] * log1p(  Î½[k]/( Î± + Râ‚‚[k] + v[2]) ) for v in F[k] ] ) )
    for k in 1:m
        l += cont_incr(k)
        if náµ‰[k] > 0
            l += disc_incr(k)
        end
    end
    return l
end
```

```julia
# Breslow estimator for cumulative hazard in frequentist Cox models
function BreslowEstimCoxModel(c::Vector{Float64},data::DataRegreNTR)
    n = data.n
    T = data.T
    Î´ = data.Î´
    Z = data.Z
    f(c,x) = exp(c'*x)
    r = [ f(c,Z[i]) for i in 1:n ] # frequencies of exact bservations
    R = cumsum( r[end:-1:1] )[end:-1:1]
    return cumsum( [ (Î´[i]==1) ? 1.0./R[i] : 0.0 for i in 1:n] ) 
end
```
